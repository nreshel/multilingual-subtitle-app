```python
# Necessary imports for the notebook
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the CSV file
csv_file = 'features_all_languages.csv'
df = pd.read_csv(csv_file)

# Display the first few rows of the DataFrame
df.head()
```

```python
# Show the index and columns of the DataFrame
print("\nIndex of the DataFrame:")
print(df.index)

print("\nColumns of the DataFrame:")
print(df.columns)
```

```python
# Basic Information about the DataFrame
display("Basic Information about the DataFrame:")
df.info()

# Check for missing values
display("\nMissing Values:")
display(df.isnull().sum())

# Summary Statistics
print("\nSummary Statistics:")
display(df.describe())
```

```python
# Check for duplicated rows
print("Checking for Duplicates:")
duplicates = df.duplicated()
print(f"Number of duplicate rows: {duplicates.sum()}")

# Drop duplicate rows if any
df_cleaned = df.drop_duplicates()
print(f"Number of rows after dropping duplicates: {len(df_cleaned)}")

# Transpose the DataFrame and check for null values in columns
df_transposed = df_cleaned.T
print("\nTransposed DataFrame:")
print(df_transposed.head())

# Check for null values in the transposed DataFrame
print("\nNull values in transposed DataFrame columns:")
print(df_transposed.isnull())

```

```python
# Count the number of samples in each label group
print("Counting samples in each label group:")
label_group_counts = df_cleaned['label'].value_counts()
print(label_group_counts)

# Plot the count of samples in each label group
plt.figure(figsize=(10, 6))
sns.barplot(x=label_group_counts.index, y=label_group_counts.values, palette="viridis")
plt.xlabel('Language')
plt.ylabel('Number of Samples')
plt.title('Count of Samples by Language')
plt.xticks(rotation=45)
plt.show()
```

```python
# Split data by labels into different DataFrames
label_dfs = {label: df_cleaned[df_cleaned['label'] == label].reset_index(drop=True) for label in df_cleaned['label'].unique()}
display(label_dfs)

# Save each label-specific DataFrame to a CSV file
for label, df_label in label_dfs.items():
    filename = f"{label}_data.csv"
    df_label.to_csv(filename, index=False)
    print(f"Saved DataFrame for label '{label}' to '{filename}'")

# Print out the number of samples in each label-specific DataFrame and display the first few rows
for label, df_label in label_dfs.items():
    print(f"\nDataFrame for label: {label}")
    print(f"Number of samples: {len(df_label)}")
    display(df_label.head())  # Using display for better formatting in Jupyter Notebook
```

```python
# Plot histograms for each feature
df.hist(bins=30, figsize=(20, 15))
plt.suptitle("Feature Distributions")
plt.show()
```

```python
# Plot the heatmap for all numerical values in the dataset
numerical_df = df.drop(columns=['label', 'filename'])
plt.figure(figsize=(20, 15))
sns.heatmap(numerical_df.corr(), annot=True, cmap='coolwarm')
plt.title('Heatmap of All Numerical Values')
plt.show()
```
